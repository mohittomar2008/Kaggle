# Filling in Missing values automatically

```python
df.head(5)
```

```python
# get a small subset of the NFL dataset

subset_nfldata= df.loc[ :,'EPA':'Season' ].head()
subset_nfldata
```

```python
subset_nfldata.isnull().sum()
```

```python
# replace missing values with 0

subset_nfldata.fillna(0, inplace=True)
```

```python
subset_nfldata.isnull().sum()
```

```python
# replace all NA's the values that comes directly after it in the same column

subset_nfldata.fillna(method='bfill', axis=0).fillna(0)
```
