```python
import pandas as pd
import numpy as np
```

```python
from scipy import stats
from sklearn.preprocessing import minmax_scale
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

```python
# take 1000 random numbers

orignal_data= np.random.exponential(size=1000)
```

```python
scaled_data= minmax_scale(orignal_data)
```

```python
sns.displot(orignal_data, kind='kde')
```

```python
sns.displot(scaled_data, kind='kde')
```

```python
# Normalization
```

```python
normalised_data= stats.boxcox(orignal_data)
```

```python
sns.displot(normalised_data, kind='kde')
```
